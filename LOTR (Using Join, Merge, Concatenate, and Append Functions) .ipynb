```python
df2 = pd.read_csv(r"C:\Users\esraa\Desktop\DataSets\Lord of the Rings\LOTR 2.csv")
df2
```

# Merge Function

```python
#### inner join (Returns records that have matching values in both dataframe) ####

#df1.merge(df2)
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
#### Full Outer Join (Returns all records when there is a match in either left or right dataframe) ####

df1.merge(df2, how = 'outer')
```

```python
#### Left Outer Join (Returns all records from the left dataframe(df1), and the matched records from the right dataframe(df2)) ####

df1.merge(df2, how = 'left')
```

```python
#### Right Outer Join (Returns all records from the right dataframe(df2), and the matched records from the left dataframe(df1)) ####

df1.merge(df2, how = 'right')
```

```python
#### Cross join (combines each row from the first dataframe with each row from the second dataframe) ####

df1.merge(df2, how = 'cross')
```

# Join Function

```python
#### on: specify what column do we want to join on ####
#### how: join type (inner join, full outer join, left outer join, right outer join) ####
#### lsuffix: suffix for column in left dataframe(df1) ####
#### rsuffix: suffix for column in right dataframe(df2) ####

df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
#### Join on index ####

df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right')
df3
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')
df3
```

# Concatenate Function
Putting one dataframe on top of the other

```python
pd.concat([df1, df2])
```

```python
#### join = 'inner' :  ONLY take the columns that are the same in the 2 dataframes, and concatenate these columns together ####

pd.concat([df1, df2], join = 'inner')
```

```python
#### join = 'outer' : take all the columns from the 2 dataframes, and concatenate these columns together ####
#### default (axis = 0) ####

pd.concat([df1, df2], join = 'outer')
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```

# Append Function
Used to append rows from one dataframe to the end of another dataframe

```python
df1.append(df2)
```
